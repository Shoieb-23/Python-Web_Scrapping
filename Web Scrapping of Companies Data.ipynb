```python

url=rqst.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(1),headers=headers).text
```

```python
soup=bsp(url,'lxml')
```

```python
#print(soup.prettify())
```

```python
comp_of_india=soup.find_all('h1') [0]
```

```python
heading=comp_of_india.text.strip()
```

```python
List_comp=soup.find_all('h2',class_="companyCardWrapper__companyName")
```

```python
companies_scrapped=[i.text.strip() for i in List_comp]
```

```python
Rating=soup.find_all('span', class_='companyCardWrapper__companyRatingValue')
```

```python
Rating_scrapped=[i.text.strip() for i in Rating]
```

```python
Reviews_scrapped=[i.text.strip() for i in (soup.find_all('span', class_='companyCardWrapper__ActionCount'))]
```

```python
Div=soup.find_all('div',class_='companyCardWrapper__tertiaryInformation')

data={
    'Names':companies_scrapped,
    'Reviews':Rating_scrapped
    
}

df=pd.DataFrame(data)

```

```python
import re

ages_=[]
for i in soup.find_all('span',class_='companyCardWrapper__interLinking'):
    m=re.search(r'(\d+ years old)',i.text.strip())
   
    if  m:
        age = m.group(0)
        ages_.append( age)

for i in ages_:
    df3.loc[len(df3)]=i
```

```python
 extracted_data = []
for i in range(len(Rating)):
    wi=[k. text. strip() for k in soup.find_all('div',class_='companyCardWrapper__tertiaryInformation')[i].
                find_all('span',class_='companyCardWrapper__ActionCount')]
    extracted_data.append(wi)
    
    
          
   
```

```python
df2=pd.DataFrame(extracted_data,columns=[['R','Salaries','Interviews','Jobs','Benefits','Photos']])

scrapped_df = pd.concat([df, df2],axis=1)
scrapped_df.columns=[['Names', 'Review','Ratings', 'Salaries','Interviews','Jobs','Benefits','Photos']]
scrapped_df=pd.concat([scrapped_df,df3],axis=1)
```

```python
scrapped_df.columns=[['Names','Review','Ratings','Salaries',
'Interviews','Jobs','Benefits','Photos','Old']]
```
